```python

# remove space
def remove_space(text):
    """
    remove extra spaces and ending space if any
    """
    spaces = ['\u200b', '\u200e', '\u202a', '\u202c', '\ufeff', '\uf0d8', '\u2061', '\x10', '\x7f', '\x9d', '\xad', '\xa0']
    for space in spaces:
        text = str(text).replace(space, ' ')
    text = text.strip()
    text = re.sub('\s+', ' ', text)
    return text


"""
clean data, remove special characters and unwanted sapces , newline, tabs
and store into clean_address
"""

def special_char_removal(space_removed):
    #for i in dayDF['uniquewords']:
    newLineRemoved = str(space_removed).replace('\\n',' ')
    newTabRemoved = str(newLineRemoved).replace('\\t',' ')
    clean_lst = re.sub('[^A-Za-z0-9]',' ', str(newTabRemoved))
    return clean_lst


def remove_largeNumbers(sentence):
    remove_tokens = []
    tokens = sentence.lower().split()
    for i in tokens:
        if len(i) >=0 and not re.match('[^\d]', i):
            remove_tokens.append(i)
    tokens = [e for e in tokens if e not in tuple(remove_tokens)]
    flat_lst_large_nos = " ".join(tokens)
    return flat_lst_large_nos


def remove_single_char(sentence):
    new_tokens =[]
    tokens = sentence.lower().split()
    for i in tokens:
        tok =re.sub(r"\b[a-zA-Z]\b", "", str(i))
        new_tokens.append(tok)
    flat_lst_special_tok = " ".join(new_tokens)
    return flat_lst_special_tok

    
```

```python
def cleaning(sentence):
    space_removed = remove_space(sentence)
    #print(space_removed)
    special_char_removed = special_char_removal(space_removed)
    #print(special_char_removed)
    space_removed = remove_space(special_char_removed)
    #print(space_removed)
    big_nos_removed = remove_largeNumbers(space_removed)
    #print(big_nos_removed)
    single_char_removed = remove_single_char(big_nos_removed)
    #print(single_char_removed)
    space_removed = remove_space(single_char_removed)

    return space_removed

```

```python
clean_text = []
for i in data['catalogueinfo']:
    cleaned = cleaning(i)
    clean_text.append(cleaned)
data['cleaned_catalogue_info'] = clean_text
```
