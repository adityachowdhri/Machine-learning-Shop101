# Problem Statement :

Given dataset of two columns :
1.No of years of experience 
2.Salary

Try to predict salary for given year of experience using linear regression.

Linear regression : y = mx + c.


```python
# import all the lib
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# read the dataset using pandas
data = pd.read_csv('Salary_Data.csv')
```

```python
# Provides some information regarding the columns in the data
data.info()
```

```python
X = data["YearsExperience"]
Y = data['Salary']

mean_x = np.mean(X)
mean_y = np.mean(Y)
m = len(X)

# y intercept and gradient

numer = 0
denom = 0
for i in range(m):
    numer += (X[i]-mean_x)*(Y[i]-mean_y)
    denom += (X[i]-mean_x)**2
    b1 = numer/denom
    b0 = mean_y - (b1*mean_x)

# These Plots help to explain the values and how they are scattered and related to each other

max_x = np.max(X) 
min_x = np.min(X)


x = np.linspace(min_x, max_x, 10)
y = b0 + b1*x
print(b0,b1)
plt.plot(x,y, color = "#58b970" , label = "regression line")
plt.scatter(X,Y, c = "#ef5423", label = "scatter plot")

#plt.figure(figsize=(12,6))
#sns.pairplot(data,x_vars=['YearsExperience'],y_vars=['Salary'],size=4,kind='scatter')
#plt.xlabel('Years')
#plt.ylabel('Salary')
#plt.title('Salary Prediction')
plt.legend()
plt.show()
rsq = r2_score(y_test,y_pred)
print(rsq)
```

```python
# Import Segregating data from scikit learn (Helps to split data into two parts : 
#1) Train data - used for model training 2) Test data - Model evaluation
from sklearn.model_selection import train_test_split
```

```python
# Split the data for train and test 
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.7,random_state=100)
```

```python
# Create new axis for x column
X_train = X_train[:,np.newaxis]
X_test = X_test[:,np.newaxis]
```

```python
# Importing Linear Regression model from scikit learn
from sklearn.linear_model import LinearRegression
```

```python
# Fitting the model
lr = LinearRegression()
lr.fit(X_train,y_train)
```

```python
# Predicting the Salary for the Test values
y_pred = lr.predict(X_test)
```

```python
# Importing metrics for the evaluation of the model
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# calculate Mean square error
mse = mean_squared_error(y_test,y_pred)
```

```python
# Calculate R square vale
rsq = r2_score(y_test,y_pred)
```

```python
print('mean squared error :',mse)
print('r square :',rsq)
```

```python
# Intecept and coeff of the line
print('Intercept of the model:',lr.intercept_)
print('Coefficient of the line:',lr.coef_)
```
