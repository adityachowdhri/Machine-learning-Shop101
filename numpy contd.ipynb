## reorganizing arrays

```python
before = np.array([[1,2,3,4],[5,6,7,8]])
print(before)

after= before.reshape((2,2,2))
print(after)
```

```python
#vertically stacking vectors
v1 = np.array([1,2,3,4])
v2 = np.array([5,6,7,8])
np.vstack([v1,v2,v1])
```

```python
# horizontal stack
h1 = np.zeros((2,4))
h2 = np.full((2,4),2)
np.hstack([h1,h2])
```
